```python
from retinaface import RetinaFace
import cv2
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

model = RetinaFace.build_model()

def detect_and_save_faces(input_image_path, output_folder, face_idx):
    # Load the input image
    image = cv2.imread(input_image_path)

    # Perform face detection using RetinaFace
    # detector = RetinaFace(quality="normal")
    print("loading retinaface..")
    # faces = RetinaFace.extract_faces(input_image_path, align=False, allow_upscaling=False)
    faces = RetinaFace.detect_faces(input_image_path, threshold=0.9, model=model,)
    print("retinaface loaded.", faces)
    
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop through each detected face and save it as a 300x300 PNG file
    for i, face_info in enumerate(faces.values()):
        # print(i, face_info)
        x, y, w, h = map(int, face_info["facial_area"])
        print(face_info["facial_area"])
        face_region = image[y:y+h, x:x+w]

        # Resize the face region to 300x300
        print("resizing..")
        face_resized = cv2.resize(face_region, (300, 300))
        
        face_idx = face_idx + 1
        # Save the face as a PNG file in the output folder
        output_path = os.path.join(output_folder, f"face_{face_idx}.png")
        # face_rgb = cv2.cvtColor(face_resized, cv2.COLOR_BGR2RGB)
        
        cv2.imwrite(output_path, face_resized)
        print(output_path, " writed.")

    print(f"{len(faces)} faces detected and saved to {output_folder}")
    return face_idx


# test
# input_image_path = '01.png'
# output_folder = 'output/'
# detect_and_save_faces(input_image_path, output_folder, 0)
    
import os
path_img = "images"
path_faces = "faces"

for gender  in os.listdir(path_img):
    path_gender = os.path.join(path_img, gender)
    for name in os.listdir(path_gender):
        path_name = os.path.join(path_img, gender, name)
        face_idx = 0
        for path_file in os.listdir(path_name):
            path_img_file = os.path.join(path_name, path_file)
            print(path_img_file)
            # input_image_path = '01.png'
            output_folder = os.path.join(path_faces, gender, name)
            face_idx = detect_and_save_faces(path_img_file, output_folder, face_idx)
            
```

```python
from deepface import DeepFace
import cv2
import os
import numpy as np
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

def detect_and_save_faces(input_image_path, output_folder):
    # Load the input image
    # image = cv2.imread(input_image_path)

    # Detect faces using the SSD model
    # detector_backend='retinaface' 'mtcnn'
    detected_faces = DeepFace.extract_faces(input_image_path,  detector_backend='mtcnn', enforce_detection = False)
    

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop through each detected face and save it as a 300x300 PNG file
    for i, face in enumerate(detected_faces):
        face = face['face']
        # print(face)
        
        face_rgb = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
        frame_normed = 255 * (face_rgb - face_rgb.min()) / (face_rgb.max() - face_rgb.min())
        face2 = np.array(frame_normed, np.int)
        
        # import matplotlib.pyplot as plt
        # plt.imshow(face2)
        # plt.show()
        
        
        
        # Save the face as a PNG file in the output folder
        output_path = os.path.join(output_folder, f"face_{i}.png")
        
        cv2.imwrite(output_path, face2)

    print(f"{len(detected_faces)} faces detected and saved to {output_folder}")


input_image_path = '01.png'
output_folder = 'output/'
detect_and_save_faces(input_image_path, output_folder)
```
